```python
import numpy as np
```

```python
a = np.array([23, 44, 67])
b = np.array([[22,33,44], [66,77,88]])
```

```python
print(a.shape)
print(b.shape)
print(a.ndim)
print(b.ndim)
print(a.size)
print(b.size)
print(a.dtype)
print(b.dtype)
```

```python
x = np.array([4,5,6])
y = np.array([12,11,10])

array_sum = x + y
print(array_sum)
```

```python
array_mult = x * y
print(array_mult)
```

```python
array_sq = x ** 2
print(array_sq)
```

```python
array_addtwo = x + 2
print(array_addtwo)
```

```python
print(np.vstack([x,y]))
print(np.hstack([x,y]))
```

```python
mat1 = np.hstack([x,y])
print(mat1)
```

```python
print(np.mean(mat1))
print(np.median(mat1))
print(np.std(mat1))
print(np.min(mat1))
print(np.max(mat1))
```
